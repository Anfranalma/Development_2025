```python
def odds_even_chars(text):
    even = []
    odds = []
    for i in range(len(text)):
        if i % 2 == 0:
            even.append(text[i])
        else:
            odds.append(text[i])
    print(''.join(even) + ' ' + ''.join(odds))
    
odds_even_chars('Angelo')
            
```
