```python
import pandas as pd
datos = pd.read_csv("celcius.csv")
```

```python
datos.head()
```

```python
datos.info()
```

```python
import seaborn as sb

sb.scatterplot(x="celcius", y ="fahrenheit", data=datos, hue="fahrenheit",palette="coolwarm")
```

```python
#separation por caracteristicas (x)
x=datos["celcius"]
y=datos["fahrenheit"]
```

```python
y
```

```python
x.values
```

```python
x_procesada = x.values.reshape(-1,1)
y_procesada = y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x_procesada, y_procesada)

```

```python
model.predict([[8]])
model.predict([[123]])
model.predict([[1024]])
```

```python
model.score(x_procesada, y_procesada)
```
